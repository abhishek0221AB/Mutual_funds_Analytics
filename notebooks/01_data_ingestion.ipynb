```python
import pandas as pd

fund_master = pd.read_csv("../data/raw/fund_master.csv")

print("=== COLUMNS IN FUND MASTER ===")
print(fund_master.columns.tolist())

print("\n=== SHAPE ===")
print(fund_master.shape)

print("\n=== HEAD ===")
print(fund_master.head())
```

```python
print("=== UNIQUE FUND HOUSES ===")
print(fund_master["fund_house"].unique())
print("Total:", fund_master["fund_house"].nunique())
```

```python
print("=== UNIQUE CATEGORIES ===")
print(fund_master["category"].unique())

print("\n=== UNIQUE SUB-CATEGORIES ===")
print(fund_master["sub_category"].unique())
```

```python
print("=== RISK GRADES ===")
print(fund_master["risk_category"].value_counts())
```

```python
print("=== AMFI CODE STRUCTURE ===")
print("Sample codes :", list(fund_master["amfi_code"].head(5)))
print("Code length  :", fund_master["amfi_code"].astype(str).str.len().unique())
print("Code range   :", fund_master["amfi_code"].min(), "→", fund_master["amfi_code"].max())
```

```python
nav_history = pd.read_csv("../data/raw/nav_history.csv")

master_codes = set(fund_master["amfi_code"].unique())
nav_codes    = set(nav_history["amfi_code"].unique())
missing      = master_codes - nav_codes

if not missing:
    print("All AMFI codes in fund_master exist in nav_history")
else:
    print(f"{len(missing)} codes missing in nav_history:")
    for code in sorted(missing):
        name = fund_master[fund_master["amfi_code"] == code]["scheme_name"].values[0]
        print(f"   • {code} — {name}")
```

```python
print("=== DATA QUALITY SUMMARY — Day 1 ===")
print(f"fund_master  → Rows: {len(fund_master)} | Nulls: {fund_master.isnull().sum().sum()} | Dupes: {fund_master.duplicated().sum()}")
print(f"nav_history  → Rows: {len(nav_history)} | Nulls: {nav_history.isnull().sum().sum()} | Dupes: {nav_history.duplicated().sum()}")
print(f"AMFI Codes   → {' All matched' if not missing else f' {len(missing)} missing'}")
```
